```python
#Block of code creates an empty board and a function to get all legal moves on such a board
#Then create a function to check wheter a match is won or not. 
def emptyBoard():
    return [0,0,0,0,0,0,0,0,0]

def getAllLegalMoves(board):
    L=list()
    for i in range(9):
        if board[i]==0:
            L.append(i)
    return L

#for i in range(10):
 #   board=numpy.random.choice([0,1,2],(9))
  #  print(board,getAllLegalMoves(board))

def isWinningBoard(board):
    if sum(board)==0:
        return False
    if board[0]==board[1] and board[1]==board[2] and board[0] != 0:
        return True
    if board[0]==board[4] and board[4]==board[8] and board[0] != 0:
        return True
    if board[6]==board[4] and board[4]==board[2] and board[6] != 0:
        return True
    if board[3]==board[4] and board[4]==board[5] and board[3] != 0:
        return True
    if board[6]==board[7] and board[7]==board[8] and board[6] != 0:
        return True
    if board[0]==board[3] and board[3]==board[6] and board[0] != 0:
        return True
    if board[1]==board[4] and board[4]==board[7] and board[1] != 0:
        return True
    if board[2]==board[5] and board[5]==board[8] and board[2] != 0:
        return True
    #...
    return False
```

```python
#create class PolicyBoard and methods to define its policies for boards
class PolicyBoard:
    def __init__(self, board):
        self.board = board

        legalMoves = len(getAllLegalMoves(board))

        policy=numpy.ones(legalMoves)/legalMoves


        self.policy = policy

board=emptyBoard()
board[0] = 1
board[3] = 1
p1 = PolicyBoard(board)

print(p1.policy)
```

```python
#AI- start vs perfect 

def AI_vs_perfect(loops, bp_reset, board_p):
    import pickle
    #perfect_strat = pickle.load(numpy.fromfile('perfectPolicy.p'))
    with open("perfectPolicy.p","rb") as file_handle:
        perfect_strat = pickle.load(file_handle)


    perfectPolicy = list(perfect_strat.items())

    #Learning rate säger hur mycket policyn ska ändras efterb varje körning

    learning_rate = 0.1

    C=[".","O","X"]
    
    if bp_reset == True:
        board_policies = []
    else:
        board_policies = board_p
    
    # Variabler för att hålla koll på hur policyn gör ifrån sig
    score = [0]

    results=[]
    wins = 0
    losses = 0
    draws = 0
    #antalet gånger koden ska köras, dvs hur många gånger tränar vi policyn
    for loops in range(loops):

        game_won = False
        board=emptyBoard()
        currentPlayer=1

        track_boards=[]
        track_choices=[]

        while not game_won:

            all_legal_moves = getAllLegalMoves(board)

            if len(all_legal_moves) != 0:
                if currentPlayer == 2:
                    for policy in perfectPolicy:
                        if policy[0] == tuple(board):
                            perfectP = policy[1]


                    choice = random.choice(all_legal_moves, p=perfectP)     

                elif currentPlayer == 1:  

                    board_policy = None

                    for bp in board_policies:
                        if bp.board == board:
                            board_policy = bp             

                    if board_policy == None:
                        board_policies.append(PolicyBoard(copy.deepcopy(board)))
                        board_policy = board_policies[-1] 

                    choice = random.choice(all_legal_moves, p=board_policy.policy)

                    track_boards.append(copy.deepcopy(board))
                    track_choices.append(choice)

                if choice in all_legal_moves:

                    #insert 1 or 2 into board at location choice
                    board[choice]=currentPlayer

                    game_won = isWinningBoard(board)

                    if game_won:
                        #showBoard(board)
                        #print("Game won by player: ", currentPlayer)

                        if currentPlayer == 1:
                            score.append(score[-1] + 1)
                            wins += 1

                            for i in range(len(track_choices)):
                                for bp in board_policies:
                                    if bp.board == track_boards[i]:
                                        for j in range( len(getAllLegalMoves(bp.board)) ):
                                            if getAllLegalMoves(bp.board)[j] == track_choices[i]:
                                                bp.policy[j] += learning_rate
                                                bp.policy/=bp.policy.sum()

                                                break



                        else:
                            losses += 1
                            score.append(score[-1] -1)
                            for i in range(len(track_choices)):
                                for bp in board_policies:
                                    if bp.board == track_boards[i]:
                                        for j in range( len(getAllLegalMoves(bp.board)) ):
                                            if getAllLegalMoves(bp.board)[j] == track_choices[i]:
                                                if bp.policy[j] - learning_rate > 0:
                                                    bp.policy[j] -= learning_rate
                                                    bp.policy/=bp.policy.sum()
                                                break


                    currentPlayer=3-currentPlayer

                else:
                    print("Incorrect move, try again!")
            else:    
                #print("Game ended in a draw")
                draws += 1
                score.append(score[-1] + 0.5)
                game_won = True
                break

    print("won", wins, " times")
    print("lost", losses, " times")
    print("drew", draws, " times")

    plot(score)
```
